```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
# Initialise PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.Aldi_db
collection = db.milk
```

```python
# URL of page to be scraped
url = 'https://www.aldi.com.au/en/groceries/fresh-produce/dairy-eggs/'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
response
```

```python
results = soup.find_all('div',class_="ratio-container m-ratio-57x25 box--description m-no-ratio-on-phone")

# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('div',class_="box--description--header").text
        # Identify and return price of listing
        price = result.find('span', class_='box--value').text
        # Identify and return link to listing
#         link = result.a['href']

        # Run only if title, price, and link are available
        if (title and price):
            # Print results
            print('-------------')
            print(title)
            print(price)
#             print(link)

            # Dictionary to be inserted as a MongoDB document
            post = {
                'title': title,
                'price': price,
#                 'url': link
            }

            collection.insert_one(post)

    except Exception as e:
        print(e)
```
