
### Average milk price 

```python
file_two = "aldimilk.csv"
df2 = pd.read_csv(file_two, encoding="ISO-8859-1")
df2.head(50)
```

```python
df2.dtypes
```

```python

df3=df2.drop(df2.index[23:414])
```

```python
df3
```

```python
df3['price'] = df3.price.str.replace('per litre,''?' ,'')
df3['price'] = df3.price.str.replace(' per Litre,''?' ,'')
df3['price'] = df3['price'].str.replace('$', '')

df3['price'] = df3.price.str.replace('80c per 100g,''?' ,'')
```

```python
df3
```

```python
df3.dropna()
```

```python
df3['price'] = pd.to_numeric(df3['price'])
```

```python
average3= df3["price"].mean()
average3
```
