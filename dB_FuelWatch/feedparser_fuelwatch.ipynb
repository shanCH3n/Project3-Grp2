## Convert RSS feed to Pandas DF

```python
import pandas as pd
```

```python
stationdata1 = {}
for entry in f.entries:
    stationdata1.setdefault('location', [])
    stationdata1.setdefault('price', [])
    stationdata1.setdefault('address', [])
    stationdata1.setdefault('updated', [])
    stationdata1.setdefault('latitude', [])
    stationdata1.setdefault('longitude', [])
    stationdata1['location'].append(entry.location)
    stationdata1['price'].append(entry.price)
    stationdata1['address'].append(entry.address)
    stationdata1['updated'].append(entry.updated)
    stationdata1['latitude'].append(entry.latitude)
    stationdata1['longitude'].append(entry.longitude)
```

```python
df = pd.DataFrame(stationdata1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# convert data types to numeric
df['price'] = pd.to_numeric(df['price'])
```

```python
# Get mean price at every suburb
location_mean = df.groupby(['location']).mean()
```
