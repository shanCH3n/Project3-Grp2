# Using feedparser to scrape a RSS Feed and store it into MongoDB

```python
#from flask import Flask, render_template, redirect
#from flask_pymongo import PyMongo
import feedparser
import pymongo
```

```python
# Initialise PyMongo to work with MongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.fuel_db
collection = db.basicstations
```

```python
# url of page to be scraped
url = "https://www.fuelwatch.wa.gov.au/fuelwatch/fuelWatchRSS?"
```

```python
# parse url with feedparser package
f = feedparser.parse(url)
```

```python
for entry in f.entries:
    try:
        price = entry.price
        location = entry.location
        address = entry.address
        latitude = entry.latitude
        longitude = entry.longitude
        updated = entry.updated
        
        if (price and location and address and latitude and longitude and updated):
            
            
            station = {
                'price': price,
                'suburb': location,
                'address': address,
                'latitude': latitude,
                'longitude': longitude,
                'date': updated
            }
            
            collection.insert_one(station)
            
    except Exception as e:
        print(e)
    
```

```python
# Display items in MongoDB collection
extractions = db.basicstations.find()

for extract in extractions:
    print(extract)
```
